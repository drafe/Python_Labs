2. Написать скрипт, который выводит на экран «True», если элементы программно задаваемого списка представляют собой возрастающую последовательность, иначе – «False».

```python
def is_increasing(*lst):
    f = lst[0][0]
    for s in lst[0][1:]:
        if s < f:
            return False
        f = s
    return True
```

```python
a1 = list(range(0, 10, 2))
a2 = list(range(10, 0, -2))
a3 = [-1, 2, 0]
is_increasing(a3)
```

7. Напишите скрипт, который обрабатывает список строк-адресов следующим образом: сначала определяет, начинается ли каждая строка в списке с префикса «www». Если условие выполняется, то скрипт должен вставить в начало этой строки префикс «http://», а затем проверить, что строка заканчивается на «.com». Если у строки другое окончание, то скрипт должен вставить в конец подстроку «.com». В итоге скрипт должен вывести на консоль новый список с измененными адресами. Используйте генераторы списков.

```python
def internet(l_address):
    l_return = ["http://" + c + ".com"
                if (c.startswith("www") and not c.endswith(".com"))
                else "http://" + c if c.startswith("www")
                else c for c in l_address]
    return l_return
```

```python
address = ["www.youtube", "https://www.youtube123.com/", "http://www.youtube12347"]
print(internet(address))
```

8. Напишите скрипт, генерирующий случайным образом число n в
диапазоне от 1 до 10000. Скрипт должен создать массив из n целых
чисел, также сгенерированных случайным образом, и дополнить
массив нулями до размера, равного ближайшей сверху степени двойки.
Например, если в массиве было n=100 элементов, то массив нужно
дополнить 28 нулями, чтобы в итоге был массив из 28=128 элементов
(ближайшая степень двойки к 100 – это число 128, к 35 – это 64 и т.д.).

```python
def get_list_power_2(number):
    import random
    lst1 = [random.randint(0, 10000) for i in range(number)]
    i = 1
    while i < number:
        i *= 2
    lst2 = [0 for i in range(i - number)]
    
    return lst1 + lst2
```

```python
import random

n = random.randint(1, 10000)
print(n)
print(get_list_power_2(n))
```

структуру данных для хранения купюр разного достоинства в заданном
количестве. При вводе пользователем запрашиваемой суммы денег,
скрипт должен вывести на консоль количество купюр подходящего
достоинства. Если имеющихся денег не хватает, то необходимо
напечатать сообщение «Операция не может быть выполнена!».
Например, при сумме 5370 рублей на консоль должно быть выведено
«5* 1000 + 3* 100 + 1* 50 + 2* 10».

```python
ATM = {1000: 10, 500: 10, 100: 10, 50: 10, 10: 10}

money = input("Введите сумму : ")
if (not money.isdigit() or int(money) % 10 != 0):
    print("Ошибка ввода! Сумма должна быть целым числом, кратным 10!")
else:
    money = int(money)
    cash = ""
    for a in ATM:
        count = money // a if money // a <= ATM[a] else ATM[a]
        if count > 0:
            if not cash == "":
                cash += " + " + str(a) + "*" + str(count)
            else:
                cash += str(a) + "*" + str(count)
            money -= a * count
        if money == 0:
            break
    if money > 0:
        print("Операция не может быть выполнена")
    else:
        print(cash)
```

10. Напишите скрипт, позволяющий определить надежность вводимого
пользователем пароля. Это задание является творческим: алгоритм
определения надежности разработайте самостоятельно.

```python
def password_entropy(passw):
    import math

    alpha = "abcdefghijklmnopqrstuvwxyz"
    upper = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    upper_punct = "~`!@#$%^&*()-_+="
    digits = "1234567890"

    f_alpha = False
    f_upper = False
    f_upper_punct = False
    f_digits = False
    f_other = False

    for l in passw:
        if l in alpha:
            f_alpha = True
        elif l in upper:
            f_upper = True
        elif l in upper_punct:
            f_upper_punct = True
        elif l in digits:
            f_digits = True
        else:
            f_other = True

    count = 0
    if f_alpha:
        count += 26
    if f_digits:
        count += 10
    if f_upper:
        count += 26
    if f_upper_punct:
        count += 16
    if f_other:
        count += 17

    return print("Ентропия пароля равна", len(passw)*math.log(count, 2))
```

```python
pas = input("Введите пароль ")
password_entropy(pas)
```

11. Напишите генератор frange как аналог range() с дробным шагом.

Пример вызова:
    
``for x in frange(1, 5, 0.1):
    print(x)``
    
выводит ``1 1.1 1.2 1.3 1.4 … 4.9``

```python
def frange(start=0, end=1, step=0.1):
    i = start
    while i < end:
        yield i#round(i,2)
        i = round(i + step, 2)

for x in frange():
    print(x)
```

12. Напишите генератор get_frames(), который производит «оконную декомпозицию» сигнала: на основе входного списка генерирует набор списков – перекрывающихся отдельных фрагментов сигнала размера size со степенью перекрытия overlap.

Пример вызова:
``for frame in get_frames(signal, size=1024, overlap=0.5):
    print(frame)``

```python
def get_frames(signal, size=1024, overlap=0.5):
    step = round(size * overlap)
    start = 0
    for i in range(0, len(signal),step):
        yield signal[i:i+step] if i + step < len(signal) else signal[i:] 

signal = list(range(15))
for frame in get_frames(signal, size=10, overlap=0.6):
    print(frame)
 
```

13. Напишите собственную версию генератора enumerate под названием extra_enumerate. В переменной cum хранится накопленная сумма на момент текущей итерации, в переменной frac – доля накопленной суммы от общей суммы на момент текущей итерации. 

Пример вызова:

``for i, elem, cum, frac in extra_enumerate(x):
    print(elem, cum, frac)``
    
Например, для списка x=[1,3,4,2] вывод будет таким:
``(1, 1, 0.1) (3, 4, 0.4) (4, 8, 0.8) (2, 10, 1)``

```python
def extra_enumerate(x):
    s_a = sum(x)
    for i in range(0, len(x)):
        s = sum(x[:i+1])
        yield i, x[i], s, s / s_a
```

```python
x = [1, 3, 4, 2]
for i, elem, cum, frac in extra_enumerate(x):
    print(elem, cum, frac)
```

14. Напишите декоратор non_empty, который дополнительно проверяет списковый результат любой функции: если в нем содержатся пустые строки или значение None, то они удаляются. 

``@non_empty
def get_pages():
return ['chapter1', '', 'contents', '', 'line1']``

```python
import functools


def non_empty(func):
    @functools.wraps(func)
    def _wrapper(*args, **kwargs):
        # пустота есть falsy
        lst = list(filter(lambda s: s, func(*args, **kwargs)))
        return lst
    return _wrapper


@non_empty
def get_pages():
    return ['chapter1', '', 'contents', '', 'line1']

print(get_pages())
```

15. Напишите параметризированный декоратор pre_process, который осуществляет предварительную обработку (цифровую фильтрацию)списка по алгоритму: s[i] = s[i]–a∙s[i–1]. Параметр а можно задать в коде (по умолчанию равен 0.97). 

``@pre_process(a=0.93)
def plot_signal(s):
    for sample in s:
        print(sample)``

```python
def pre_process(a=0.97):
    def _decor(func):
        def _wrapp(*args, **kwargs):
            s = [args[0][i] - a * args[0][i - 1] for i in range(0, len(args[0]))]
            func(s, **kwargs)
        return _wrapp
    return _decor


@pre_process(a=0.93)
def plot_signal(s):
    for sample in s:
        print(sample)

sig = [10, 20, 30, 40, 50]
plot_signal(sig)
```

16. Напишите скрипт, который на основе списка из 16 названий
футбольных команд случайным образом формирует 4 группы по 4
команды, а также выводит на консоль календарь всех игр (игры
должны проходить по средам, раз в 2 недели, начиная с 14 сентября
текущего года). Даты игр необходимо выводить в формате «14/09/2016,
22:45». Используйте модули random и itertools

```python
import datetime
import random
import itertools

this_moment = datetime.datetime.now()
start_moment = datetime.datetime(this_moment.year, 9, 14, 22, 45)
team_list = ["Австрия", "Бельгия", "Болгария", "Венгрия",
             "Греция", "Германия", "Дания", "Италия",
             "Ирландия", "Испания", "Кипр", "Люксембург",
             "Латвия", "Литва", "Мальта", "Нидерланды"]
random.shuffle(team_list)
team_split = [[0 for j in range(4)] for i in range(4)]
for i in range(0, 4):
    team_split[i] = team_list[4*i:4*i+4]
[print("Группа №", i + 1, team_split[i]) for i in range(0, 4)]
print("\n")
match_list = [[0 for j in range(4)] for i in range(6)]
for i in range(4):
    match_list[i] = [j for j in itertools.combinations(team_split[i], 2)]
for i in range(1, 25):
    print("Игра #", i, match_list[(i - 1) // 6][(i - 1) % 6], start_moment.strftime("%d/%m/%Y %H:%M"))
    start_moment += datetime.timedelta(days=14)
```
